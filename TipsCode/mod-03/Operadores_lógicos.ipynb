```python
print('### Bem-vindo ao parque de diverses! ###')
altura = int(input('Qual  a sua altura em cm? '))

if altura >= 120:
    idade = int(input('Qual  a sua idade? '))
    if idade <= 12:
        conta = 5
        print('O ingresso vai custar R$ 5,00.')
    elif idade > 12 and idade <= 18:
        conta = 7
        print('O ingresso vai custar R$ 7,00.')
    elif idade > 18 and idade <= 44:
        conta = 12
        print('O ingresso vai custar R$ 12,00.')
    elif idade >= 45 and idade <= 55: # 45 <= idade <= 55
        conta = 0
        print('Que bom lhe ver aqui, hoje seu ingresso  gratuito!')
    else:
        conta = 24
        print('O ingresso vai custar R$ 24,00.')

    foto = input('Deseja tirar uma foto? (S/N) ') # S para sim, N para no
    if foto.upper() == 'S': # upper() converte a resposta para maiscula, assim o usurio pode responder com 's' ou 'S'
        conta += 3

    print(f'Foto adicionada ao ingresso.\nO valor do ingresso mais a foto  de R$ {conta:.2f}.')
else:
    print('No vende o ingresso!')
```
