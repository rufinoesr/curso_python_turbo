Calculadora IMC

IMC abaixo de 18.5 (não incluso), imprima "abaixo do peso";

IMC acima ou igual a 18.5 (inclusive) ou abaixo de 25.0 (não incluso), imprima "peso normal";

IMC abaixo de 25.0 (inclusive), imprima "Acima do peso.

```python
print('CALCULADORA DE IMC:')

peso = float(input('Qual é o seu peso em Kg? '))
altura = float(input('Qual é a sua altura em metros? '))

imc = peso / (altura ** 2)
print(f'Seu IMC é: {imc:.2f}.')

if imc < 18.5:
    print('Você está abaixo do peso!')

elif imc >= 18.5 and imc < 25.0: # ou elif 18.5 <= imc < 25.0:
    print('Você está com peso normal!')

else:
    print('Você está acima do peso!')
```

Código de entrega de Pizza e calcula automaticamente a conta para o usuário com base nas opções abaixo:

Opções:

Tabela de Pizza: (1) Pequena: R$ 15,00; (2) Média: R$ 20,00 e (3) Grande: R$ 25,00.

Adicionais: Peperoni na pizza pequena: + R$ 2,00; na pizza média e grande: + R$ 3,00 (S/N).

Queijo extra em todos os tamanhos: + R$ 1,00 (S/N).

```python
print('### Bem-vindo ao aplicativo de pedido de Pizza! ###')

tamanho = input('Qual tamanho de pizza você deseja? (P/M/G) ').title()

if tamanho.upper() == 'P': # ou if tamanho in ['P', 'p']:
    conta = 15
    print('Você escolheu a pizza pequena (P), o valor é de R$ 15,00.')
elif tamanho.upper() == 'M':
    conta = 20
    print('Você escolheu a pizza média (M), o valor é de R$ 20,00.')
else:  
    conta = 25 #
    print('Você escolheu a pizza grande (G), o valor é de R$ 25,00.')

adicional = input('Deseja adicional de Peperoni? (S/N) ')

if adicional.upper() == 'S':
    if tamanho.upper() == 'P':
        conta += 2
        print('Adicionado Peperoni à sua pizza + R$ 2,00!')
    else:
        conta += 3
        print('Adicionado Peperoni à sua pizza + R$ 3,00!')
else:
    print('Nenhum adicional de Peperoni.')

queijo = input('Deseja queijo extra? (S/N) ')

if queijo.upper() == 'S':
    conta += 1
    print('Adicionado queijo extra à sua pizza + R$ 1,00!')
else:
    print('Nenhum adicional de Queijo.')

print(f'A conta final da sua Pizza de tamanho \'{tamanho}\' é de R$ {conta:.2f}.')
```
