```python
altura = int(input('Qual é a sua altura em cm? '))

if altura >= 120:
    print('Vende o ingresso!')
    idade = int(input('Qual é a sua idade? '))
    if idade <= 12:
        conta = 5
        print('O ingresso vai custar R$ 5,00.')
    elif idade <= 18:
        conta = 12
        print('O ingresso vai custar R$ 12,00.')
    else:
        conta = 24
        print('O ingresso vai custar R$ 24,00.')

    foto = input('Deseja tirar uma foto? (S/N) ') # S para sim, N para não
    if foto.upper() == 'S': # upper() converte a resposta para maiúscula, assim o usuário pode responder com 's' ou 'S'
        conta += 3

    print(f'Foto adicionada ao ingresso.\nO valor do ingresso mais a foto é de R$ {conta:.2f}.')
else:
    print('Não vende o ingresso!')
```
