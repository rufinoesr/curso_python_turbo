Projeto Gorjeta:

Código que fará um cálculo de gorjeta após o fechamento da conta em um restaurante.

Solocitará o valor da conta a ser paga, o valor da porcentagem (10, 12 ou 15)%.

Solicitará o número de pessoas que dividiram a conta.

```python
conta = float(input('Informe o valor final da conta a ser paga:' ))
porcentagem = int(input('Informe a porcentagem que deseja dar de gorjeta? 10%, 12% ou 15%?')) / 100
pessoas = int(input('Informe quantas pessoas irão dividir esta conta?' ))

conta_Com_gorjeta = (conta + (conta * porcentagem))

valor_final_pessoas = conta_Com_gorjeta / pessoas

print(f'Cada pessoa deve pagar: R$ {valor_final_pessoas:.2f}.')
```
