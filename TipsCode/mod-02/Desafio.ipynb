Objetivo escrever um programa que some um número de dois dígitos.


Terá um input onde o usuário digitará um número de dois digitos e estes serão somados. Ex.: 39 onde o 3 será somado com o 9, dando resultado 12.

```python
num = input('Digite um número de dois dígitos: ') # informação passada pelo usuário é sempre do tipo string.

soma =  int(num[0]) + int(num[1]) # converte os caracteres individuais (string) de volta para inteiros (int) para realizar a soma.

print(f'A soma dos dígitos do número informado é: {soma}.')
```

Escrever um código que irá calcular em média quantas semanas de vida restam a uma pessoa aqui na Terra.

O objetivo é considerar a entrada de acordo com a sua idade.

Exemplo. Tem alguém de 15 anos, e presumasse que a pessoa viverá aqui na Terra até os 90 anos.

Então, quantas semanas ainda restam em sua vida?

```python
idade = input("Digite sua idade: ")
limite = (90 - int(idade)) * 48
print(limite)
```
