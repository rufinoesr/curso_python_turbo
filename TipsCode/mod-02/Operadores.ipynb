Operadores Matemáticos

```python
print(3 + 5) # adição.
print(8 - 2) # subtração.
print(2 * 3) # multiplicação.
print(6 / 3) # tecnicamente, esta divisão sempre retorna um float.
print(2 ** 3) # potência.
print(5 // 2) # divisão inteira.
```

PEMDAS - (Parênteses, Expoentes, Multiplicação/Divisão, Adição/Subtração)

A ordem das operações aritméticas. Parênteses têm a maior prioridade, seguidos por expoentes (**), multiplicação e divisão (*, /, //, %), e, por fim, adição e subtração (+, -), avaliados da esquerda para a direita

```python
print(3 * 3 + 3 / 3 - 2)
```

```python
print(3 * (3 + 3) / 3 - 3)
```

```python
peso = float(input('Digite seu peso em kg:  '))
altura = float(input('Digite sua altura em metros:  '))

imc = peso / (altura ** 2)

print(f'Seu IMC é: {imc:.2f}')
```

```python
peso = float(input('Digite seu peso em kg:  '))
altura = float(input('Digite sua altura em metros:  '))

imc = peso / (altura ** 2)

new_imc = int(imc)

print(f'Seu IMC é ' + str(new_imc) + '.')
```

Arredondar um número para mais ou para menos

```python
print(8 / 3) # divisão normal, retorna float.
```

```python
print(int(8 / 3)) # converte o resultado da divisão para inteiro.
```

```python
print(round(8 / 3))  # arredonda para o inteiro mais próximo.    
```

```python
print(round(8 / 3, 3)) # arredonda para três casas decimais.
```

```python
print(round(8 / 3, 2))  # arredonda para duas casas decimais.    
```

```python
print(round(8 / 3, 1)) # arredonda para uma casa decimal.
```

```python
print((8 // 3)) # arredonda para o inteiro mais próximo.
```

```python
print(type(8 // 3))
```

```python
print(type(8 / 3))
```

```python
result = 8 / 4
print(result)
```

Operadores de Atribuição Composta

```python
result = 8 / 4
result /= 2 # equivalente a result = result / 2
print(result)
```

```python
soma = 5 + 4
soma += 3  # equivalente a x = x + 3
print(soma)
```
