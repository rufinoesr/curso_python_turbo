Forma de exibir um nmero inteiro grande na tela, utilizando um sublinhado ( _ ) como separador visual para facilitar a leitura.

Float - nmeros que possuem casas decimais

```python
print(24.6 + 34.7)
```

```python
print(f'{24.6 + 34.7:.2f}') # resultado  um nmero do tipo float
```

```python
print(int(24.6) + int(34.7))
```

```python
soma = 24.6 + 34.7

new_soma = int(soma)

print(f'A soma inteira  ' + str(new_soma) + '.')
```
