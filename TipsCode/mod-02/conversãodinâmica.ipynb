Converte diretamente a entrada de texto.

Erros (ValueError): Se a string contiver letras, símbolos ou estiver vazia (ex: int("abc") ou int("")), o Python lançará um ValueError.

Essa conversão é muito comum ao usar a função input(), que sempre recebe dados como texto por padrão.

```python
idade = int(input("Digite sua idade: "))

print("Sua idade é: " + str(idade))
```

Strings representando números decimais (ex: "10.5") não podem ser convertidas diretamente por int(). Primeiro, deve-se converter para float() e depois para int(): int(float("10.5")).

```python
altura = float(input("Digite sua altura: "))

print("Sua altura é: " + str(altura) + ' cm') # concatenando a string com a variável altura convertida para string.
```

```python
altura = float(185.0)

print(int(altura)) # convertendo a variável altura para inteiro, o que resulta em 185.
```
