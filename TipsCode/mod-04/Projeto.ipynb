### Game: Pedra, Papel e Tesoura ###

### É um jogo divertido e uma ferramenta de tomada de decisão. Como virar uma moeda ou rolar dados, ajuda a resolver disputas de forma rápida e justa. Sua simplicidade e imprevisibilidade o tornam ideal para resolver pequenos conflitos. 

```python
import random


papel = '''
    _______
---'   ____)____
          ______)
          _______)
         _______)
---.__________)

'''


tesoura = '''
    _______
---'   ____)____
          ______)
       __________)
      (____)
---.__(___)


'''


pedra = '''
    _______
---'   ____)
      (_____)
      (_____)
      (____)
---.__(___)

'''
```

```python
print("### Bem-vindo ao jogo: Pedra, Papel e Tesoura! ### \n")

imagem_maos = [pedra, papel, tesoura]

jogador = int(input("Escolha: 0 para Pedra, 1 para Papel e 2 para Tesoura: "))
computador = random.randint(0, 2) #Gerando um número aleatório entre 0 e 2 para representar a escolha do computador.
print(f"Você escolheu: {imagem_maos[jogador]}") #Imprimindo a escolha do jogador, que é a imagem correspondente à escolha dele.
print(f"O computador escolheu: {imagem_maos[computador]}") #Imprimindo a escolha do computador, que é a imagem correspondente à escolha dele.
if jogador == computador:
    print("==> EMPATE!")
elif (jogador == 0 and computador == 2) or (jogador == 1 and computador == 0) or (jogador == 2 and computador == 1):
    print("==> Você GANHOU!")
else:
    print("==> Você PERDEU!")
    
```
