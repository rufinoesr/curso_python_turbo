### Jogo amigo secreto ###

Crie uma aplicação que mostre em tela nomes aleatórios.

```python
print(''' ### AMIGO SECRETO ### ''')

import random

secret_friend = ['Tiago', 'Rufino', 'Gustavo', 'Bruno', 'Rafael', 'Renan', 'Rodrigo', 'Humberto', 'Felipe', 'Miguel', 'João', 'Ewerton', 'Fernando']

nome_sorteado = random.choice(secret_friend)    

print(f'O amigo secreto sorteado foi: {nome_sorteado}')
```
