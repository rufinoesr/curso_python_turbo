Mersenne Twister

O Mersenne Twister (MT19937) é o gerador de números pseudoaleatórios (PRNG) padrão no módulo
{Link: random https://docs.python.org/pt-br/3.8/library/random.html} do Python, oferecendo alta velocidade e um longo período de
. Embora não seja seguro para criptografia, é ideal para simulações e modelagens gerais.

```python
import random

#import meu_modulo
```

```python
random_int = random.randint(1, 10) #gera um número inteiro aleatório entre 1 e 10, incluindo ambos os extremos.
print(random_int)

#print(meu_modulo.name)
```

```python
number_0_to_1 = random.random() #gera um número aleatório entre 0 e 1.
print(number_0_to_1)
```

```python
number_0_to_1 = random.random() * 10 #gera um número aleatório entre 0 e 10 sem incluir o 10.
print(number_0_to_1)
```

Intervalos semiabertos.

```python
random_float = random.uniform(1, 10) #gera um número de ponto flutuante aleatório entre 1 e 10, incluindo ambos os extremos.
print(random_float)
```

Jogo de cara ou coroa

Ao gerar números aleatórios, definindo se é cara ou coroa.

```python
print('### Jogo de Cara ou Coroa! ###')

random_int = random.randint(0, 1) #gera um número inteiro aleatório entre 0 e 1, incluindo ambos os extremos. 0 pode representar "cara" e 1 pode representar "coroa".
if random_int == 0:
    print('Cara')
else:
    print('Coroa')
```
