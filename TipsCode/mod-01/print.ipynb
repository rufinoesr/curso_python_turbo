INPUT() function

```python
print('Jesus te ama!')
input('Qual é o seu nome? ')
```

```python
print('Oi, ' + input('Qual é o seu nome? ') + '!')
```

```python
name = input('Qual é o seu nome? ')
print('Oi, ')
print(name)
```

```python
name = input('Qual é o seu nome? ')
print('Oi, ' + name) # concatenação com o +
```

```python
name = input('Qual é o seu nome? ')
print('Oi,', name) # vírgula adiciona um espaço automaticamente.
```

```python
name = input('Qual é o seu nome? ')
print('Oi, ', end='') # end='' evita a quebra de linha.
print(name)
```

```python
name = input('Qual é o seu nome? ')
print('Oi, ', name, sep='') # sep='' evita o espaço entre os argumentos.
```

```python
print('Oi, \'amigo\'!') # escape com \ para usar aspas simples ou duplas dentro de uma string delimitada por aspas simples ou duplas.
```

```python
name = input('Qual é o seu nome? ')
name = name.strip() # strip() remove espaços em branco no início e no final da string.
print(f'Oi, {name}!')
```

```python
name = input('Qual é o seu nome? ')
name = name.capitalize() # capitalize() coloca a primeira letra em maiúscula.
print(f'Oi, {name}!')
```

```python
name = input('Qual é o seu nome? ')
name = name.title() # title() coloca a primeira letra de cada palavra em maiúscula.
print(f'Oi, {name}!')
```
