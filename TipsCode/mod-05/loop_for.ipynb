# Criar um algoritmo usando loop for para percorrer todos os elementos da lista e retorne a soma.

```python
notas_alunos = [7.5, 8.0, 6.0, 9.5, 5.0, 8.5, 7.0, 6.5, 9.0, 8.0]
sun = 0
for nota in notas_alunos:
    sun += nota # Adiciona cada nota à variável 'sun' para calcular o total.
print(f'Total das notas: {sun}')
print(max(notas_alunos)) # Imprime a maior nota da lista de notas dos alunos.
```

# Criar um algoritmo usando loop for pra descobrir qual é o maior valor dessa lista.

```python
notas_alunos = [7.5, 8.0, 6.0, 9.5, 5.0, 8.5, 7.0, 6.5, 9.0, 8.0]
maior_nota = 0 # Inicializa a variável 'maior' com a primeira nota da lista.
for nota in notas_alunos:
    if nota > maior_nota:
        maior_nota = nota # Atualiza a variável 'maior' se a nota atual for maior.
print(f'Maior nota: {maior_nota}')
```
