# Função RANGE

```python
print(range(1,10))
```

```python
for number in range(1,10): # O loop for itera sobre a sequência de números gerada por range(1,10), que inclui os números de 1 a 9.
    print(number)
```

```python
for i in range(1,11,3):
    print(i)
```

# Soma de 1 até 100:

```python
total = 0
for i in range(1,101):
    total += i # O operador += é uma forma abreviada de escrever total = total + i, ou seja, ele adiciona o valor de i à variável total a cada iteração do loop.
print(total)
```
