Programa que calcula a idade que uma pessoa pode se aposentar, considerando que ela atinja 65 anos de idade para ter o benefício do INSS.

```python
print(''' ### Bem-vindo(a) ao programa do INSS ### ''')
print('Para saber se tem direito ao benefício do INSS, faremos algumas perguntas:')

nome = input('Informe seu nome: ')
idade = int(input('Informe sua idade: '))

if idade >= 65:
    print(f'{nome}, você possiu a idade necessária para receber o benefício do INSS.')
elif idade < 65:
    print(f'{nome}, você ainda não atingiu a idade necessária de 65 anos para receber o benefício do INSS.')
    ano = (2026 - idade) + 65
    beneficio = 65 - idade
    print(f'Ainda faltam {beneficio} anos e somente em {ano} você se aposentará e passará a receber merecidamente sua aposentadoria.')

    
```
