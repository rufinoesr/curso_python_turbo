### Escreva um programa para ler um valor do teclado e apresentar o seu antecessor.

```python
num = int(input('Digite um número e descubra seu antecessor: '))

antecessor = num - 1

print(f'O antecessor do {num} é {antecessor}.')
```

### Escreva um programa para ler as dimensões de um retângulo (base e altura).
### Calcule e apresente a área. 

```python
altura = float(input('Digite a altura do retângulo em metros: '))
base = float(input('Digite a base do retângulo em metros: '))

area = altura * base

print(f'A área do retângulo é {area} metros quadrados.')
```
