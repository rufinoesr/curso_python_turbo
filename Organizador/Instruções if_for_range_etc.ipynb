Ferramentas de controle de fluxo

Instruções if

```python
x = int(input('Digite um número inteiro: '))

if x < 0:
    x = 0
    print('Negativo convertido para zero')

elif x == 0:
    print('Zero')
elif x == 1:
    print('Um')
else:
    print('Mais')
```

 Instruções for

A instrução for do Python itera sobre os itens de qualquer sequência (seja uma lista ou uma string), na ordem que aparecem na sequência.
