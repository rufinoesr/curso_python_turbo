Desafio: Calcular o preço por litro

```python
produto = input('Digite o nome do produto: ')
peso = float(input('Digite o peso do produto (em ml/g): '))
preço = float(input('Digite o preço do produto: R$ '))

preço_por_unidade = preço / peso
print(f'O preço por ml/g do produto {produto} é: R$ {preço_por_unidade:.2f} por ml/g')
```

```python
peso1 = float(input('Digite o peso do produto (em ml/g): '))
peso2 = float(input('Digite o peso do produto (em ml/g): '))

preço1 = float(input('Digite o preço do produto: R$ '))
preço2 = float(input('Digite o preço do produto: R$ '))

preço_por_unidade1 = preço1 / peso1
preço_por_unidade2 = preço2 / peso2

if preço_por_unidade1 < preço_por_unidade2:
    print(f'O produto mais barato é o primeiro, com preço R$ {preço1:.2f}')
elif preço_por_unidade2 < preço_por_unidade1:
    print(f'O produto mais barato é o segundo, com preço R$ {preço2:.2f}')
else:
    print('Os dois produtos têm o mesmo preço.')
```
