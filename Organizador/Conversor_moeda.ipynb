# Conversor de moeda:

```python
import customtkinter
from pegar_moedas import nomes_moedas, nomes_conversoes

customtkinter.set_appearance_mode('dark')
customtkinter.set_default_color_theme('dark-blue')
janela = customtkinter.CTk()
janela.geometry('500x500')

# get the available currencies
moeda_lista = nomes_moedas()
# build a simple conversions map: each currency can convert to all other currencies
dic_conversoes_disponiveis = {code: [c for c in moeda_lista.keys() if c != code] for code in moeda_lista}

titulo = customtkinter.CTkLabel(janela, text='Conversor de Moedas!', font=('Arial', 20))
texto_moeda_origem = customtkinter.CTkLabel(janela, text='Selecione a moeda de origem:')
texto_moeda_destino = customtkinter.CTkLabel(janela, text='Selecione a moeda de destino:')

def carregar_moedas_destino(moeda_selecionada):
    lista_moedas_destino = dic_conversoes_disponiveis.get(moeda_selecionada, [])
    campo_moeda_destino.configure(values=lista_moedas_destino)

campo_moeda_origem = customtkinter.CTkOptionMenu(janela, values=list(dic_conversoes_disponiveis.keys()), command=carregar_moedas_destino)
campo_moeda_destino = customtkinter.CTkOptionMenu(janela, values=[])

def carregar_moedas_destino(moeda_origem):
    moedas_destino = dic_conversoes_disponiveis.get(moeda_origem, [])
    campo_moeda_destino.configure(values=moedas_destino)

def converter_moeda():
    moeda_origem = campo_moeda_origem.get()
    moeda_destino = campo_moeda_destino.get()
    print(f'Convertendo de {moeda_origem} para {moeda_destino}...')

botao_converter = customtkinter.CTkButton(janela, text='Converter', command=converter_moeda)

lista_moedas = customtkinter.CTkScrollableFrame(janela)

for codigo_moeda in moeda_lista:
    label_moeda = customtkinter.CTkLabel(lista_moedas, text=f'{codigo_moeda}: {moeda_lista[codigo_moeda]}')
    label_moeda.pack()


titulo.pack(padx=10, pady=10)
texto_moeda_origem.pack(padx=10, pady=3)
campo_moeda_origem.pack(padx=10, pady=3)
texto_moeda_destino.pack(padx=10, pady=3)
campo_moeda_destino.pack(padx=10, pady=5)
botao_converter.pack(padx=10, pady=5)
lista_moedas.pack(padx=10, pady=10, fill='both', expand=True)


janela.mainloop()
```
