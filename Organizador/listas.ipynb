# Sequência de Fibonacci:

```python
# a soma de dois elementos define a próxima

a, b = 0, 1
while a < 10:
    print(a)
    a, b = b, a + b
```

```python
a, b = 0, 1
while b < 10:
    print(b)
    b, a = a, b + a
```

O laço de repetição (while) executa enquanto a condição (aqui: a < 10) permanece verdadeira.
Em Python, como em C, qualquer valor inteiro que não seja zero é considerado verdadeiro; zero é considerado falso.
